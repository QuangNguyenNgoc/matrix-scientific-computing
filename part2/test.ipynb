## 1. Phép toán Chéo hóa (Diagonalization)

---
### Test Case 1: Normal (Ma trận vuông không đối xứng, các trị riêng phân biệt)
* **Input:**
    $$A = \begin{pmatrix} 62.5 & -18.2 \\ 108.3 & -37.1 \end{pmatrix}$$
* **Expected Value:**
    * **Trị riêng (Eigenvalues):** $\lambda_1 \approx 35.292$, $\lambda_2 \approx -9.892$.
    * **Ma trận đường chéo $D$:**
        $$D = \begin{pmatrix} 35.292 & 0 \\ 0 & -9.892 \end{pmatrix}$$

```python
demo_diagonalize([
    [62.5, -18.2],
    [108.3, -37.1]
])
```

---
### Test Case 2: Boundary (Ma trận đối xứng, trị riêng lặp)giá trị $0$ (ma trận suy biến).
* **Input:**
    $$A = \begin{pmatrix} 2 & -1 & -1 \\ -1 & 2 & -1 \\ -1 & -1 & 2 \end{pmatrix}$$
* **Expected Value:**
    * **Trị riêng:** $\lambda_1 = 3$, $\lambda_2 = 3$ (bội 2) và $\lambda_3 = 0$.
    * **Ma trận đường chéo $D$:**
        $$D = \begin{pmatrix} 3 & 0 & 0 \\ 0 & 3 & 0 \\ 0 & 0 & 0 \end{pmatrix}$$


```python
demo_diagonalize([
    [2.0, -1.0, -1.0],
    [-1.0, 2.0, -1.0],
    [-1.0, -1.0, 2.0]
])
```

---
### Test Case 3: Error / Exception (Ma trận khiếm khuyết - Defective Matrix)
* **Input:**
    $$A = \begin{pmatrix} 5 & 4 & 2 \\ 0 & 5 & 1 \\ 0 & 0 & 5 \end{pmatrix}$$
* **Expected Value:**
    * **Trị riêng:** $\lambda = 5$ (bội 3).
    * **Không gian vector riêng:** Số chiều của không gian vector riêng tương ứng với $\lambda = 5$ nhỏ hơn 3.
    * Hệ thống phải bắt được ngoại lệ (Exception) hoặc trả về cờ trạng thái `False`

```python
try:
    demo_diagonalize([
        [5.0, 4.0, 2.0],
        [0.0, 5.0, 1.0],
        [0.0, 0.0, 5.0]
    ])
except Exception as e:
    print(f"Lỗi: {type(e).__name__} -> {e}")
```

---
## 2. Phân rã Giá trị Suy biến (SVD - Singular Value Decomposition)

---
### Test Case 1: Normal (Ma trận chữ nhật, Full Hạng)
* **Input:**
    $$A = \begin{pmatrix} 12.5 & 8.2 \\ 8.1 & 12.7 \\ 8.0 & -8.5 \end{pmatrix}$$
* **Expected Value:**
    * **Kích thước output:** $U$ là ma trận $3 \times 3$, $\Sigma$ là ma trận $3 \times 2$, và $V^T$ là ma trận $2 \times 2$.
    * **Giá trị suy biến (Singular Values):** Phải tồn tại 2 giá trị suy biến khác không ($\sigma_1, \sigma_2 > 0$) nằm trên đường chéo chính của $\Sigma$, được sắp xếp giảm dần ($\sigma_1 \ge \sigma_2$).

```python
demo_svd([
    [12.5, 8.2],
    [8.1, 12.7],
    [8.0, -8.5]
])
```

---
### Test Case 2: Boundary (Ma trận hạng hụt / Rank-deficient)
* **Input:**
    $$A = \begin{pmatrix} 2.5 & -5.0 & 7.5 \\ 5.0 & -10.0 & 15.0 \\ -2.5 & 5.0 & -7.5 \end{pmatrix}$$
* **Expected Value:**
    * **Hạng của ma trận:** $r = 1$.
    * **Giá trị suy biến:** Chỉ có duy nhất 1 giá trị $\sigma_1 > 0$. Các giá trị còn lại $\sigma_2 \approx 0$ và $\sigma_3 \approx 0$.

```python
demo_svd([
    [2.5, -5.0, 7.5],
    [5.0, -10.0, 15.0],
    [-2.5, 5.0, -7.5]
])
```

---
### Test Case 3: Error / Edge (Dữ liệu đầu vào bất hợp lệ)
* **Input:**
    $$A = \begin{pmatrix} 1.4 & \text{NaN} \\ \text{Inf} & 3.2 \end{pmatrix}$$
* **Expected Value:**
    * Hệ thống không được lặp vô hạn hay xử lý ngầm và ra kết quả sai. Bắt buộc phải văng ngoại lệ báo lỗi liên quan đến giá trị bất hợp lệ.

```python
try:
    demo_svd([
        [1.4, math.nan],
        [math.inf, 3.2]
    ])
except Exception as e:
    print(f"Lỗi: {type(e).__name__} -> {e}")
```
