# Test Cases: Chéo hóa Ma trận và Phân rã SVD

Đưa các ma trận thử nghiệm vào hai hàm cốt lõi `demo_diagonalize` và `demo_svd` nhằm rà soát tính ổn định của hệ thống.

Mỗi phần bao gồm **5 test case** bao phủ: thường, biên và lỗi.

```python
import math
from decomposition import demo_svd
from diagonalization import demo_diagonalize
```

---
## 1. Phép toán Chéo hóa (Diagonalization)

---
### Test Case 1: Ma trận 2×2 không đối xứng, trị riêng phân biệt
* **Input:**
    $$A = \begin{pmatrix} 62.5 & -18.2 \\ 108.3 & -37.1 \end{pmatrix}$$
* **Expected Value:**
    * **Trị riêng (Eigenvalues):** $\lambda_1 \approx 35.26$, $\lambda_2 \approx -9.86$.
    * **Sai số tái cấu trúc** $\|A - PDP^{-1}\|_{\max} < 10^{-4}$.

```python
demo_diagonalize([
    [62.5, -18.2],
    [108.3, -37.1]
])
```

---
### Test Case 2: Ma trận 3×3 đối xứng, trị riêng phân biệt
* **Input:**
    $$A = \begin{pmatrix} 4 & 1 & 0 \\ 1 & 3 & 1 \\ 0 & 1 & 2 \end{pmatrix}$$
* **Expected Value:**
    * Ba trị riêng phân biệt (ma trận đối xứng luôn chéo hóa được).
    * **Sai số tái cấu trúc** $\|A - PDP^{-1}\|_{\max} < 10^{-4}$.

```python
demo_diagonalize([
    [4, 1, 0],
    [1, 3, 1],
    [0, 1, 2]
])
```

---
### Test Case 3: Ma trận đường chéo 4×4
* **Input:**
    $$A = \begin{pmatrix} 5 & 0 & 0 & 0 \\ 0 & -3 & 0 & 0 \\ 0 & 0 & 7 & 0 \\ 0 & 0 & 0 & 1 \end{pmatrix}$$
* **Expected Value:**
    * $P = I_4$, $D = A$ (ma trận đường chéo tự đã chéo hóa).
    * **Sai số tái cấu trúc** $\|A - PDP^{-1}\|_{\max} < 10^{-4}$.

```python
demo_diagonalize([
    [5, 0, 0, 0],
    [0, -3, 0, 0],
    [0, 0, 7, 0],
    [0, 0, 0, 1]
])
```

---
### Test Case 4: Ma trận 3×3 đối xứng, trị riêng lặp kèm giá trị 0 (suy biến)
* **Input:**
    $$A = \begin{pmatrix} 2 & -1 & -1 \\ -1 & 2 & -1 \\ -1 & -1 & 2 \end{pmatrix}$$
* **Expected Value:**
    * **Trị riêng:** $\lambda_1 = \lambda_2 = 3$, $\lambda_3 = 0$.
    * Ma trận vẫn chéo hóa được vì đây là ma trận đối xứng.
    * **Sai số tái cấu trúc** $\|A - PDP^{-1}\|_{\max} < 10^{-4}$.

```python
demo_diagonalize([
    [2.0, -1.0, -1.0],
    [-1.0, 2.0, -1.0],
    [-1.0, -1.0, 2.0]
])
```

---
### Test Case 5: Ma trận khiếm khuyết, không chéo hóa được
* **Input:**
    $$A = \begin{pmatrix} 5 & 4 & 2 \\ 0 & 5 & 1 \\ 0 & 0 & 5 \end{pmatrix}$$
* **Expected Value:**
    * **Trị riêng:** $\lambda = 5$ (bội 3), không gian riêng chỉ có 1 chiều.
    * Hệ thống phải ném `ValueError` do ma trận không đủ vector riêng độc lập.

```python
try:
    demo_diagonalize([
        [5.0, 4.0, 2.0],
        [0.0, 5.0, 1.0],
        [0.0, 0.0, 5.0]
    ])
except Exception as e:
    print(f"Lỗi: {type(e).__name__} -> {e}")
```

## 2. Phân rã Giá trị Suy biến (SVD - Singular Value Decomposition)

---
### Test Case 1: Ma trận 3×2 chữ nhật có nhiều dòng hơn cột
* **Input:**
    $$A = \begin{pmatrix} 12.5 & 8.2 \\ 8.1 & 12.7 \\ 8.0 & -8.5 \end{pmatrix}$$
* **Expected Value:**
    * $U \in \mathbb{R}^{3 \times 3}$, $\Sigma \in \mathbb{R}^{3 \times 2}$, $V^T \in \mathbb{R}^{2 \times 2}$.
    * Hai giá trị kỳ dị $\sigma_1 \geq \sigma_2 > 0$.
    * **Sai số tái cấu trúc** $\|A - U\Sigma V^T\|_{\max} < 10^{-4}$.

```python
demo_svd([
    [12.5, 8.2],
    [8.1, 12.7],
    [8.0, -8.5]
])
```

---
### Test Case 2: Ma trận vuông 3×3, full hạng
* **Input:**
    $$A = \begin{pmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 10 \end{pmatrix}$$
* **Expected Value:**
    * Ba giá trị kỳ dị $\sigma_1 \geq \sigma_2 \geq \sigma_3 > 0$.
    * **Sai số tái cấu trúc** $\|A - U\Sigma V^T\|_{\max} < 10^{-4}$.

```python
demo_svd([
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 10]
])
```

---
### Test Case 3: Ma trận 2×4 có nhiều cột hơn dòng
* **Input:**
    $$A = \begin{pmatrix} 1 & 2 & 3 & 4 \\ 5 & 6 & 7 & 8 \end{pmatrix}$$
* **Expected Value:**
    * $U \in \mathbb{R}^{2 \times 2}$, $\Sigma \in \mathbb{R}^{2 \times 4}$, $V^T \in \mathbb{R}^{4 \times 4}$.
    * Tối đa 2 giá trị kỳ dị khác 0 (hạng $\leq \min(m,n) = 2$).
    * **Sai số tái cấu trúc** $\|A - U\Sigma V^T\|_{\max} < 10^{-4}$.

```python
demo_svd([
    [1, 2, 3, 4],
    [5, 6, 7, 8]
])
```

---
### Test Case 4: Ma trận hạng hụt
* **Input:**
    $$A = \begin{pmatrix} 2.5 & -5.0 & 7.5 \\ 5.0 & -10.0 & 15.0 \\ -2.5 & 5.0 & -7.5 \end{pmatrix}$$
* **Expected Value:**
    * Hạng của ma trận $r = 1$.
    * Chỉ có $\sigma_1 > 0$; $\sigma_2 \approx \sigma_3 \approx 0$.
    * **Sai số tái cấu trúc** $\|A - U\Sigma V^T\|_{\max} < 10^{-4}$.

```python
demo_svd([
    [2.5, -5.0, 7.5],
    [5.0, -10.0, 15.0],
    [-2.5, 5.0, -7.5]
])
```

---
### Test Case 5: Ma trận chứa giá trị NaN/Inf không hợp lệ
* **Input:**
    $$A = \begin{pmatrix} 1.4 & \text{NaN} \\ \text{Inf} & 3.2 \end{pmatrix}$$
* **Expected Value:**
    * Hệ thống phải ném `ValueError` báo lỗi giá trị bất hợp lệ, không được xử lý ngầm.

```python
try:
    demo_svd([
        [1.4, math.nan],
        [math.inf, 3.2]
    ])
except Exception as e:
    print(f"Lỗi: {type(e).__name__} -> {e}")
```
